# **Scenario 1: Money at a fixed rate for an unsecured purchase (e.g. credit card) for an individual.**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the datasets
delinquency_rates = pd.read_csv('delinquency_rate_on_credit_card_loans.csv', parse_dates=['Date'])
inflation_rates = pd.read_csv('inflation_consumer_prices.csv', parse_dates=['Date'])
credit_card_interest_rates = pd.read_csv('commercial_bank_interest_rate_on_credit_card_plans.csv', parse_dates=['Date'])
unemployment_rates = pd.read_csv('unemployment_rate.csv', parse_dates=['Date'])
total_credit_card_debt = pd.read_csv('credit_card_debt.csv', parse_dates=['Date'])
```

## **Delinquency Rates**

Delinquency rates offer a clear picture of how often people are falling behind on payments. If this rate is climbing, it's a heads-up to the team that customers might be struggling and that they may need to brace for more unpaid debts.

```python
delinquency_rates.head(10)
```

```python
# Plotting Delinquency Rate
plt.figure(figsize=(14, 7))
plt.plot(delinquency_rates['Date'], delinquency_rates['Rate'])
plt.title('Graphic 1: Delinquency Rate on Credit Card Loans Over Time')
plt.xlabel('Year')
plt.ylabel('Delinquency Rate (%)')

plt.figtext(0.5, 0.01, "* This graph shows the trend of credit card loan delinquency rates, which is crucial for understanding changing credit risk.",
            wrap=True, horizontalalignment='center', fontsize=10)

plt.show()
```

## **Inflation Rates**

Rising inflation often means costs for goods and services are going up. This affects how much money people have to spend on repaying debts. For the team, this can signal when to adjust interest rates to stay profitable.

```python
inflation_rates.tail(10)
```

```python
# Plotting Inflation Rates Over Time
plt.figure(figsize=(14, 7))
plt.plot(inflation_rates['Date'], inflation_rates['Index'])
plt.title('Graphic 2: Inflation Rates Over Time')
plt.xlabel('Year')
plt.ylabel('Inflation Rate (%)')

plt.figtext(0.5, 0.01, "* This graph displays the changes in inflation rates over time, providing insights into economic trends and the purchasing power of money.",
            wrap=True, horizontalalignment='center', fontsize=10)

plt.show()
```

## **Interest Rate on Credit Card**

These rates tell us how much it costs for customers to borrow money. If these are high and still people are racking up debt, it indicates that customers aren't deterred by costlier credit, which could lead to higher profits but also greater risk of default.

```python
# Plotting Commercial Bank Interest Rate on Credit Card Plans
plt.figure(figsize=(14, 7))
plt.plot(credit_card_interest_rates['Date'], credit_card_interest_rates['Rate'])
plt.title('Graphic 3: Interest Rate on Credit Card Plans Over Time')
plt.xlabel('Year')
plt.ylabel('Interest Rate (%)')

plt.figtext(0.5, 0.01, "* Rising credit card interest rates can indicate a tightening in credit conditions and an increased cost of borrowing.",
            wrap=True, horizontalalignment='center', fontsize=10)

plt.show()
```

## **Unemployment Rates**

Unemployment rates can predict how much disposable income people might have. Higher unemployment usually means less spending and borrowing, which can influence the team's decisions on credit limits and risk management.

```python
# Plotting Unemployment Rates Over Time
plt.figure(figsize=(14, 7))
plt.plot(unemployment_rates['Date'], unemployment_rates['Rate'])
plt.title('Graphic 4: Unemployment Rates Over Time')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')

plt.figtext(0.5, 0.01, "* The unemployment rate trend is a vital indicator of the labor market's health, influencing consumer spending and overall economic activity.",
            wrap=True, horizontalalignment='center', fontsize=10)

plt.show()
```

## **Total Credit Card Debt**

Monitoring total credit card debt helps the team gauge overall consumer spending and borrowing habits. If debt levels are up, it might mean people feel financially stable enough to spend, but it could also signal future repayment issues if the debt grows too quickly.

```python
# Plotting Total Credit Card Debt Over Time

plt.figure(figsize=(14, 7))
plt.plot(total_credit_card_debt['Date'], total_credit_card_debt['Debt'])
plt.title('Graphic 5: Total Credit Card Debt Over Time')
plt.xlabel('Year')
plt.ylabel('Total Debt (in USD)')

plt.figtext(0.5, 0.01, "* This graph illustrates the trends in total credit card debt, shedding light on consumer borrowing patterns and financial health.",
            wrap=True, horizontalalignment='center', fontsize=10)

plt.show()
```

## **Economic Indicators**

By having all these indicators in one view, the team can spot patterns and trends they might miss if they looked at each data point in isolation.

```python
import matplotlib.pyplot as plt

# Create a single figure
plt.figure(figsize=(14, 7))

# Create primary axis
ax1 = plt.gca()

# Plot graphs with percentage values on primary y-axis
ax1.plot(delinquency_rates['Date'], delinquency_rates['Rate'], label='Delinquency Rate (%)', color='blue')
ax1.plot(inflation_rates['Date'], inflation_rates['Index'], label='Inflation Rate (%)', color='green')
ax1.plot(credit_card_interest_rates['Date'], credit_card_interest_rates['Rate'], label='Credit Card Interest Rate (%)', color='red')
ax1.plot(unemployment_rates['Date'], unemployment_rates['Rate'], label='Unemployment Rate (%)', color='purple')

# Set labels for primary y-axis
ax1.set_xlabel('Year')
ax1.set_ylabel('Percentage (%)')

# Create secondary axis for monetary data
ax2 = ax1.twinx()
ax2.plot(total_credit_card_debt['Date'], total_credit_card_debt['Debt'], label='Total Credit Card Debt (Trillions USD)', color='orange')

# Set label for secondary y-axis
ax2.set_ylabel('Total Debt (Trillions USD)')

# Adding title
plt.title('Graphic 6: Economic Indicators Over Time')

# Adding legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Adding a footnote
plt.figtext(0.5, 0.01, "* This graph combines trends of various economic indicators over time.", wrap=True, horizontalalignment='center', fontsize=10)

# Show the plot
plt.show()
```

## **Correlation Matrix**


The correlation matrix is a powerful tool because it lets the team see at a glance how closely related these different financial indicators are to each other.

**Note:** It's important to remark that correlation does not imply causation, so these relationships do not necessarily indicate that one variable causes the other to change. They only indicate that there is a statistical relationship between the two. Other external factors and economic conditions could be influencing these indicators simultaneously.

```python
# Convert index to DatetimeIndex
delinquency_rates.index = pd.to_datetime(delinquency_rates['Date'])
inflation_rates.index = pd.to_datetime(inflation_rates['Date'])
credit_card_interest_rates.index = pd.to_datetime(credit_card_interest_rates['Date'])
unemployment_rates.index = pd.to_datetime(unemployment_rates['Date'])
total_credit_card_debt.index = pd.to_datetime(total_credit_card_debt['Date'])

# Resample datasets to yearly mean and select only the relevant column
delinquency_rates_yearly = delinquency_rates.resample('Y')['Rate'].mean()
inflation_rates_yearly = inflation_rates.resample('Y')['Index'].mean()
credit_card_interest_rates_yearly = credit_card_interest_rates.resample('Y')['Rate'].mean()
unemployment_rates_yearly = unemployment_rates.resample('Y')['Rate'].mean()
total_credit_card_debt_yearly = total_credit_card_debt.resample('Y')['Debt'].mean()

# Merge datasets
data = pd.concat([delinquency_rates_yearly, inflation_rates_yearly, credit_card_interest_rates_yearly, unemployment_rates_yearly, total_credit_card_debt_yearly], axis=1)
data.columns = ['Delinquency Rate', 'Inflation Rate', 'Credit Card Interest Rate', 'Unemployment Rate', 'Total Credit Card Debt']

# Calculate the correlation matrix
corr_matrix = data.corr()

# Plotting the correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Graphic 7: Yearly 5x5 Correlation Matrix')
plt.show()
```

## **Exploratory data analysis**

**Delinquency Rate and Other Factors:**

1. There's a moderate positive correlation with the Unemployment Rate (0.33), which suggests that as unemployment goes up, delinquency rates tend to increase as well. This is logical because people who are unemployed may struggle to pay off their debts.
2. There's a slight negative correlation with Inflation Rate (-0.21) and Credit Card Interest Rate (-0.12), indicating that higher delinquency rates are slightly associated with lower inflation and lower credit card interest rates, although the relationship is weak.
3. Almost no correlation with Total Credit Card Debt (-0.029), suggesting delinquency rates do not move in tandem with the total amount of credit card debt.


**Inflation Rate and Other Factors:**
1. Shows a moderate positive correlation with the Credit Card Interest Rate (0.48), which is expected because interest rates often rise with inflation to control economic growth.
2. There's a moderate negative correlation with the Unemployment Rate (-0.44), suggesting that as inflation increases, unemployment tends to decrease, which could reflect periods of economic growth.
3. A slight positive correlation with Total Credit Card Debt (0.3), indicating that as inflation increases, there is a tendency for credit card debt to rise slightly.

**Credit Card Interest Rate and Other Factors:**
1. A strong positive correlation with Total Credit Card Debt (0.87), suggesting that as credit card interest rates rise, the total credit card debt also tends to increase significantly. This might indicate that higher interest rates do not immediately deter borrowing on credit cards, or it may reflect compounding interest on existing debt.
2. A moderate negative correlation with the Unemployment Rate (-0.4), indicating that higher interest rates may be associated with lower unemployment, possibly reflecting central bank policies to cool down an overheating economy by raising rates.

**Unemployment Rate and Other Factors:**
1. Besides the correlations already mentioned, it has a slight negative correlation with Total Credit Card Debt (-0.35), which could suggest that higher unemployment may lead to a decrease in credit card debt, potentially because unemployed individuals are less likely to qualify for credit or might be paying down debt.
